```python
#1.	Load the given CSV file containing text and label columns into a Pandas DataFrame.
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/NLP/reviews_dataset.csv')

```

```python
#2.	Display the first three records of the dataset and print the total number of rows and columns.
print(df.head(3))
print(df.shape[0])
print(df.shape[1])
```

```python
#3.	Print the column names and data types of each column.
print(df.columns)
print(df.dtypes)
```

```python
#4.	Check whether the dataset contains any missing values and display the count of missing values per column.
print(df.isnull().sum())
```

```python
#5.	Find the number of unique labels present in the dataset.
unique_labels = df['label'].nunique()  #nunique() gives the count of unique values, unique() gives the unique values
print(unique_labels)
```

```python
#6.	Display the count of records for each label.
label_counts = df['label'].value_counts()
print(label_counts)
```

```python
#7.	Filter and display all rows where the label is "Negative".
negative_rows = df[df['label'] == 'Negative']
print(negative_rows)
```

```python
#8. Create a new column that stores the number of characters present in each text
df["review_char_count"] = df["review"].fillna('').astype(str).str.len()
```

```python
#9. Create another column that stores the number of words in each text.
df['word_count'] = df['review'].fillna('').apply(lambda x: len(x.split()))
```

```python
#10.	Convert all text entries to lowercase and store the result in a new column without modifying the original text column.
df['lowercase_review'] = df['review'].str.lower()
```

```python
#11.	Standardize the label column by converting all labels to lowercase.
df['label'] = df['label'].str.lower()
```

```python
#12.	Create a new column in which the labels are encoded as follows: positive → 1, neutral → 0, negative → −1.
label_mapping = {'positive': 1, 'neutral': 0, 'negative': -1}
df['label_encoded'] = df['label'].map(label_mapping)
```

```python
#13.	Sort the DataFrame in descending order based on the word count column.
df = df.sort_values(by='word_count', ascending=False)
```

```python
#14.	Display the top three longest text entries based on word count.
top_three_longest = df.head(3) [['review', 'word_count']]
print(top_three_longest)
```

```python
#15. Filter and display all texts that contain more than six words and belong to the "positive" label.
positive_long_texts = df[(df['word_count'] > 6) & (df['label'] == 'positive')]
print(positive_long_texts)
```

```python
#16.	Group the data by label and compute the average word count for each group.
average_word_count = df.groupby('label')['word_count'].mean()
print(average_word_count)
```

```python
#17.	Identify the label that has the highest average word count.
highest_avg_word_count_label = average_word_count.idxmax()
print(highest_avg_word_count_label)
```

```python
#18.	Count the number of text entries with fewer than five words for each label.
short_text_counts = df[df['word_count'] < 5].groupby('label').size()
print(short_text_counts)
```

```python
#19.	Change the DataFrame index to start from 1 instead of 0.
df.index = df.index + 1
```

```python
#20.	Rename the columns text to review_text and label to sentiment.
df = df.rename(columns={'review': 'review_text', 'label': 'sentiment'})
```

```python
#21.	Update the sentiment of a specific record (given index) from "neutral" to "positive".
df.at[5, 'sentiment'] = 'positive'  #choose record 5
```

```python
#22.	Create a new DataFrame that contains only the rows with positive sentiment.
positive_df = df[df['sentiment'] == 'positive']
print(positive_df)
```

```python
#23.	Save the cleaned and updated DataFrame to a new CSV file.
output_file_path = 'cleaned_updated_data.csv'
df.to_csv(output_file_path, index=False)
print(f"DataFrame successfully saved to {output_file_path}")
```
