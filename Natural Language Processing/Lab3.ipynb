```python
#Import necessary libraries
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
#1.	Load the dataset into a Pandas DataFrame and extract the text and label columns.
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Natural Language Processing/cleaned_updated_data.csv",usecols=['review_text', 'sentiment'])
df

documents = df['review_text']
print(f"Number of documents: {len(documents)}")
print(f"First 5 documents:\n{documents.head()}")
```

```python
#2.	Perform tokenization on all documents and store the tokens corresponding to each document.
import nltk
from nltk.tokenize import word_tokenize

# Download the 'punkt' tokenizer models required for word_tokenize
nltk.download('punkt_tab')

# Tokenize each document
tokenized_documents = [word_tokenize(doc) for doc in documents.astype(str)]

print(f"Number of tokenized documents: {len(tokenized_documents)}")
print(f"First 5 tokenized documents:\n{tokenized_documents[:5]}")
```

```python
#3.	Apply case folding by converting all tokens to lowercase.
casefold_tokens= [[token.casefold() for token in doc] for doc in tokenized_documents]
casefold_tokens
```

```python
#4.	Remove stop-words from the tokenized documents.
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')

english_stop_words = set(stopwords.words('english'))
```

Stemming-based Representation

```python
#5. Apply stemming to the filtered tokens and store the stemmed tokens separately.
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()

stemmed_tokens = [[stemmer.stem(word) for word in doc_tokens] for doc_tokens in casefold_tokens]
stemmed_tokens
```

```python
#6.	Create a new column containing the final preprocessed text after stemming by joining the stemmed tokens.
df['stemtokens'] = [' '.join(token) for token in stemmed_tokens]
df
```

```python
#7.	Construct the Bag-of-Words vocabulary using the stemmed text.
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['stemtokens'])
```

```python
#8.	Determine the size of the stemming-based vocabulary and list the top 10 most frequent words.
vocab = vectorizer.get_feature_names_out()
print(f"Size of the stemming-based vocabulary: {len(vocab)}")
print(f"Top 10 most frequent words: {vocab[:10]}")
```

```python
#9.	Represent each document using Bag-of-Words frequency vectors and generate the document–term matrix (DTM) for the stemmed text.
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['stemtokens'])
```

Lemmatization-based Representation

```python
#10. Apply lemmatization to the filtered tokens and store the lemmatized tokens separately.
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords

# Download necessary NLTK data for WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')

lemmatizer = WordNetLemmatizer()

lemmatized_documents = []
for doc_tokens in casefold_tokens:
    # Filter out stop words first (assuming 'english_stop_words' is defined from previous steps)
    filtered_tokens_after_stopwords = [word for word in doc_tokens if word not in english_stop_words]
    # Then, lemmatize the remaining tokens
    lemmatized_tokens_for_doc = [lemmatizer.lemmatize(word) for word in filtered_tokens_after_stopwords]
    lemmatized_documents.append(lemmatized_tokens_for_doc)

print(f"Number of lemmatized documents: {len(lemmatized_documents)}")
print(f"First 5 lemmatized documents:\n{lemmatized_documents[:5]}")
```

```python
#11.	Create a new column containing the final preprocessed text after lemmatization by joining the lemmatized tokens.
df['lemmtokens'] = [' '.join(token) for token in lemmatized_documents]
df
```

```python
#12. Construct the Bag-of-Words vocabulary using the lemmatized text.
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
X_lemmatized = vectorizer.fit_transform(df['lemmtokens'])
```

```python
#13.	Determine the size of the lemmatization-based vocabulary and list the top 10 most frequent words.
vocab = vectorizer.get_feature_names_out()
print(f"Size of the lemmatization-based vocabulary: {len(vocab)}")
print(f"Top 10 most frequent words: {vocab[:10]}")
```

```python
#14.	Represent each document using Bag-of-Words frequency vectors and generate the document–term matrix (DTM) for the lemmatized text.
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
X_lemmatized = vectorizer.fit_transform(df['lemmtokens'])
```

```python
#15. Select any one document and display its BoW vector from:
#• the stemming-based DTM
#• the lemmatization-based DTM

from sklearn.feature_extraction.text import CountVectorizer

# Document index to display (e.g., the first document)
doc_index = 0

# --- Stemming-based DTM ---
stem_vectorizer = CountVectorizer()
X_stemmed = stem_vectorizer.fit_transform(df['stemtokens'])
stem_vocab = stem_vectorizer.get_feature_names_out()

# Get the BoW vector for the selected document from the stemmed DTM
stem_bow_vector = X_stemmed[doc_index].toarray().flatten()

# Display the stemmed BoW vector (word: count)
print(f"\nBoW vector for document {doc_index} (stemming-based):")
stem_bow_representation = {stem_vocab[i]: count for i, count in enumerate(stem_bow_vector) if count > 0}
print(stem_bow_representation)

# --- Lemmatization-based DTM ---
lemma_vectorizer = CountVectorizer()
X_lemmatized = lemma_vectorizer.fit_transform(df['lemmtokens'])
lemma_vocab = lemma_vectorizer.get_feature_names_out()

# Get the BoW vector for the selected document from the lemmatized DTM
lemma_bow_vector = X_lemmatized[doc_index].toarray().flatten()

# Display the lemmatized BoW vector (word: count)
print(f"\nBoW vector for document {doc_index} (lemmatization-based):")
lemma_bow_representation = {lemma_vocab[i]: count for i, count in enumerate(lemma_bow_vector) if count > 0}
print(lemma_bow_representation)
```

```python
#16. Identify words that appear in most documents and words that appear in very few documents, and discuss their significance in text classification.

import numpy as np

# --- Stemming-based DTM ---
# The vectorizer and X_stemmed are already created in step 15
# stem_vectorizer, X_stemmed, and stem_vocab from previous steps

# Calculate document frequency for stemming-based DTM
# A term is present in a document if its count > 0
stem_doc_frequencies = np.sum(X_stemmed > 0, axis=0).A1

# Sort by frequency
stem_sorted_indices = stem_doc_frequencies.argsort()

print("\n--- Stemming-based Analysis ---")
print(f"Total number of documents: {X_stemmed.shape[0]}")

print("\nWords appearing in most documents (Stemming):")
for i in stem_sorted_indices[::-1][:10]: # Top 10 most frequent by document
    print(f"'{stem_vocab[i]}': appears in {stem_doc_frequencies[i]} documents")

print("\nWords appearing in very few documents (Stemming):")
for i in stem_sorted_indices[:10]: # Top 10 least frequent by document
    print(f"'{stem_vocab[i]}': appears in {stem_doc_frequencies[i]} documents")


# --- Lemmatization-based DTM ---
# The vectorizer and X_lemmatized are already created in step 15
# lemma_vectorizer, X_lemmatized, and lemma_vocab from previous steps

# Calculate document frequency for lemmatization-based DTM
lemma_doc_frequencies = np.sum(X_lemmatized > 0, axis=0).A1

# Sort by frequency
lemma_sorted_indices = lemma_doc_frequencies.argsort()

print("\n--- Lemmatization-based Analysis ---")
print(f"Total number of documents: {X_lemmatized.shape[0]}")

print("\nWords appearing in most documents (Lemmatization):")
for i in lemma_sorted_indices[::-1][:10]: # Top 10 most frequent by document
    print(f"'{lemma_vocab[i]}': appears in {lemma_doc_frequencies[i]} documents")

print("\nWords appearing in very few documents (Lemmatization):")
for i in lemma_sorted_indices[:10]: # Top 10 least frequent by document
    print(f"'{lemma_vocab[i]}': appears in {lemma_doc_frequencies[i]} documents")
```

# Discussion of significance in text classification:
Words appearing in most documents (high document frequency) are often common words (even after stop-word removal, some general words might remain) or domain-specific terms that are present across nearly all texts. While they might be useful for understanding the general topic of a corpus, they often have low understanding for differentiating between different classes (e.g., positive vs. negative sentiment) because they don't strongly associate with any particular class.

Words appearing in very few documents (low document frequency) are rare terms. They can be highly discriminative if they strongly indicate a specific class. If a word only appears once or twice, its presence might be coincidental, or it might be a misspelling. In many text classification tasks, very rare words are often removed (e.g., by setting a minimum document frequency threshold in CountVectorizer) to reduce the dimensionality of the feature space and improve model robustness, as they can lead to overfitting or computational inefficiency.
