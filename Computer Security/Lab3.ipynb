1. Rail Fence Cipher encryption
2. Rail Fence Cipher decryption


```python
def clean_text(text):
    return ''.join(c.upper() for c in text if c.isalpha()) # Remove non-alphabetic characters and convert to uppercase


def encrypt_rail_fence(text, key):
    text = clean_text(text)
    rails = ['' for _ in range(key)]

    row = 0
    direction = 1   # 1 = down, -1 = up

    for char in text:
        rails[row] += char
        if row == 0:
            direction = 1
        elif row == key - 1:
            direction = -1
        row += direction

    return ''.join(rails)


def decrypt_rail_fence(cipher, key):
    pattern = []
    row = 0
    direction = 1

    # Diagonal
    for _ in cipher:
        pattern.append(row)
        if row == 0:
            direction = 1
        elif row == key - 1:
            direction = -1
        row += direction

    # Count characters per rail
    rail_count = [0] * key
    for r in pattern:
        rail_count[r] += 1

    # Fill rails
    rails = []
    index = 0
    for count in rail_count:
        rails.append(cipher[index:index+count])
        index += count

    # Reconstruct original text
    result = []
    rail_index = [0] * key

    for r in pattern:
        result.append(rails[r][rail_index[r]])
        rail_index[r] += 1

    return ''.join(result)


while True:
    print("\n Rail Fence Cipher")
    print("1. Encrypt")
    print("2. Decrypt")
    print("3. Exit")

    choice = input("Enter choice: ")

    if choice == '1':
        text = input("Enter plaintext: ")
        key = int(input("Enter key (rails): "))
        cipher = encrypt_rail_fence(text, key)
        print("Encrypted Text:", cipher)

    elif choice == '2':
        cipher = input("Enter ciphertext: ")
        key = int(input("Enter key (rails): "))
        plain = decrypt_rail_fence(cipher, key)
        print("Decrypted Text:", plain)

    elif choice == '3':
        print("Exiting Program...")
        break

    else:
        print("Invalid Choice!")
```

1. Row Transposition Cipher Encryption
2. Row Transposition Cipher Decryption

```python
import math

# Encryption
def encrypt_row_transposition(plaintext, key):
    plaintext = plaintext.replace(" ", "")
    key_len = len(key)

    num_rows = math.ceil(len(plaintext) / key_len)
    padding_len = (num_rows * key_len) - len(plaintext)
    plaintext += "X" * padding_len

    matrix = [list(plaintext[i:i+key_len])
              for i in range(0, len(plaintext), key_len)]

    ciphertext = ""
    for k in sorted(key):
        col_idx = key.index(k)
        for row in matrix:
            ciphertext += row[col_idx]

    return ciphertext


# Decryption
def decrypt_row_transposition(ciphertext, key):
    key_len = len(key)
    num_rows = len(ciphertext) // key_len

    matrix = [['' for _ in range(key_len)]
              for _ in range(num_rows)]

    curr = 0
    for k in sorted(key):
        col_idx = key.index(k)
        for r in range(num_rows):
            matrix[r][col_idx] = ciphertext[curr]
            curr += 1

    plaintext = ""
    for row in matrix:
        plaintext += "".join(row)

    return plaintext

def get_key():
    key_input = input("Enter key numbers separated by space (example: 4 3 1 2 5): ")
    key = list(map(int, key_input.split()))
    return key

def main():
    while True:
        print("\n--- ROW TRANSPOSITION CIPHER ---")
        print("1. Encrypt")
        print("2. Decrypt")
        print("3. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            plaintext = input("Enter plaintext: ")
            key = get_key()
            cipher = encrypt_row_transposition(plaintext, key)
            print("\nCiphertext:", cipher)

        elif choice == "2":
            ciphertext = input("Enter ciphertext: ")
            key = get_key()
            plain = decrypt_row_transposition(ciphertext, key)
            print("\nDecrypted Text:", plain.rstrip("X"))

        elif choice == "3":
            print("Exiting Program...")
            break

        else:
            print("Invalid choice. Try again.")

main()
```

1. Column Transposition Cipher Encryption
2. Column Transposition Cipher Dencryption

```python
import math

def encrypt(plaintext, key):
    plaintext = plaintext.replace(" ", "")
    key_len = len(key)

    rows = math.ceil(len(plaintext) / key_len)
    padded_len = rows * key_len
    plaintext = plaintext.ljust(padded_len, 'X')

    grid = [list(plaintext[i:i+key_len]) for i in range(0, len(plaintext), key_len)]

    ciphertext = ""
    for k in sorted(range(len(key)), key=lambda k: key[k]):
        for row in grid:
            ciphertext += row[k]

    return ciphertext


def decrypt(ciphertext, key):
    key_len = len(key)
    rows = len(ciphertext) // key_len

    grid = [['' for _ in range(key_len)] for _ in range(rows)]

    sorted_key_indices = sorted(range(len(key)), key=lambda k: key[k])

    col_idx = 0
    for k in sorted_key_indices:
        for r in range(rows):
            grid[r][k] = ciphertext[col_idx]
            col_idx += 1

    plaintext = "".join(["".join(row) for row in grid])
    return plaintext


def get_key():
    print("Enter numeric key separated by spaces (example: 4 3 1 2 5):")
    key = list(map(int, input().split()))
    return key


while True:
    print("\n--- COLUMNAR TRANSPOSITION CIPHER ---")
    print("1. Encrypt")
    print("2. Decrypt")
    print("3. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        text = input("Enter plaintext: ")
        key = get_key()
        cipher = encrypt(text, key)
        print("Encrypted Text:", cipher)

    elif choice == "2":
        text = input("Enter ciphertext: ")
        key = get_key()
        plain = decrypt(text, key)
        print("Decrypted Text:", plain)

    elif choice == "3":
        print("Exiting Program...")
        break

    else:
        print("Invalid choice! Try again.")
```
