1. Hill Cipher Encryption and Decryption

```python
import numpy as np

# Define the alphabet
alphas = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"


def get_key_matrix(size):
    """Accepts user input for the key matrix"""
    key = []
    print("\nEnter key matrix:")
    for i in range(size):
        temp = [int(input(f"Enter value for {i, j}: ")) for j in range(size)]
        key.append(temp)
    return np.array(key)


def text_to_numbers(text):
    """Convert text to numerical representation"""
    return [alphas.index(char) for char in text]


def numbers_to_text(numbers):
    """Convert numbers back to text"""
    return "".join(alphas[num % 26] for num in numbers)


def pad_text(plain_text, size):
    """Pad text to be a multiple of the matrix size"""
    while len(plain_text) % size != 0:
        plain_text += "X"  # Padding with 'X'
    return plain_text


def encrypt(plain_text, key):
    """Encrypt the text using Hill Cipher"""
    size = len(key)
    plain_text = pad_text(plain_text.upper(), size)

    grams_index = text_to_numbers(plain_text)
    grams = [grams_index[i:i + size] for i in range(0, len(grams_index), size)]

    cipher_grams = []
    for gram in grams:
        result = np.dot(key, gram) % 26  # Matrix multiplication mod 26
        cipher_grams.extend(result)

    cipher_text = numbers_to_text(cipher_grams)
    return cipher_text


def decrypt(cipher_text, key):
    """Decrypt the text using Hill Cipher"""
    size = len(key)

    try:
        key_inv = np.linalg.inv(key)  # Inverse of matrix
        key_inv = np.round(key_inv * np.linalg.det(key))  # Adjusting determinant
        key_inv = key_inv.astype(int)  # Convert to int
        det_inv = pow(int(np.linalg.det(key)) % 26, -1, 26)  # Modular inverse of determinant
        key_inv = (key_inv * det_inv) % 26  # Modulo 26 inverse matrix
    except:
        print("Error: Key matrix is not invertible under modulo 26.")
        return None

    grams_index = text_to_numbers(cipher_text)
    grams = [grams_index[i:i + size] for i in range(0, len(grams_index), size)]

    plain_grams = []
    for gram in grams:
        result = np.dot(key_inv, gram) % 26
        plain_grams.extend(result)

    plain_text = numbers_to_text(plain_grams).rstrip("X")  # Remove padding
    return plain_text


def hill_cipher():
    """Main function to execute encryption and decryption"""
    plain_text = input("Enter the plain text: ").upper().replace(" ", "")
    size = int(input("Enter the size of the key matrix: "))

    key = get_key_matrix(size)

    print("\nEncryption Process:")
    cipher_text = encrypt(plain_text, key)
    print(f"Cipher Text: {cipher_text}")

    print("\nDecryption Process:")
    decrypted_text = decrypt(cipher_text, key)
    if decrypted_text:
        print(f"Decrypted Text: {decrypted_text}")

hill_cipher()
```

2. RSA Algorithm

```python
import math

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

def mod_inverse(e, phi):
    for d in range(3, phi):
        if (d * e) % phi == 1:
            return d
    raise ValueError("mod_inverse does not exist")

def rsa_demo_valid_number_message():
    print("RSA Algorithm with Valid Parameters and Number Message\n")

    # Valid primes
    p = int(input("Enter prime number p: ").strip())
    q = int(input("Enter prime number q (different from p): ").strip())

    n = p * q
    phi = (p - 1) * (q - 1)

    print(f"Chosen primes: p = {p}, q = {q}")
    print(f"Calculated n = p * q = {n}")
    print(f"Calculated φ(n) = (p-1)*(q-1) = {phi}\n")

    # Valid public exponent
    e= int(input("Enter e value: "))  # gcd(17, phi) = 1

    d = mod_inverse(e, phi)
    print(f"Chosen public exponent e = {e}")
    print(f"Calculated private exponent d = {d}")

    public_key = (e, n)
    private_key = (d, n)

    print(f"\nPublic Key (e, n): {public_key}")
    print(f"Private Key (d, n): {private_key}\n")

    # Valid numeric message
    message = int(input("Enter message")) # Must be less than n = 3233
    print(f"Original Numeric Message: {message}")

    # Encrypt
    ciphertext = pow(message, e, n)
    print(f"Encrypted Message: {ciphertext}")

    # Decrypt
    decrypted = pow(ciphertext, d, n)
    print(f"Decrypted Message: {decrypted}")

if __name__ == "__main__":
    rsa_demo_valid_number_message()
```

3. ElGamal Algorithm

```python
import math

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

def mod_inverse(a, p):
    return pow(a, p - 2, p)

def elgamal_key_generation(p, g, x):
    y = pow(g, x, p)
    print(f"\nPublic Key: (g = {g}, y = {y}, p = {p})")
    print(f"Private Key: x = {x}")
    return (g, y, p), x

def elgamal_encrypt(public_key, M, k):
    g, y, p = public_key
    C1 = pow(g, k, p)
    C2 = (M * pow(y, k, p)) % p
    print(f"\nEncryption:")
    print(f"C1 = {g}^{k} mod {p} = {C1}")
    print(f"C2 = {M} * {y}^{k} mod {p} = {C2}")
    return (C1, C2)

def elgamal_decrypt(private_key, public_key, ciphertext):
    x = private_key
    g, y, p = public_key
    C1, C2 = ciphertext
    K = pow(C1, x, p)
    K_inv = mod_inverse(K, p)
    M = (C2 * K_inv) % p
    print(f"\nDecryption:")
    print(f"K = {C1}^{x} mod {p} = {K}")
    print(f"K⁻¹ = {K_inv}")
    print(f"M = {C2} * {K_inv} mod {p} = {M}")
    return M

def main():
    print("=== ElGamal Encryption/Decryption with Valid Parameters ===")

    # Valid parameters
    p = 467       # a prime number
    g = 2         # a primitive root mod p
    x = 127       # private key (1 < x < p)
    M = 123       # message to encrypt (M < p)
    k = 89        # random number (1 < k < p)

    # Run ElGamal
    public_key, private_key = elgamal_key_generation(p, g, x)
    ciphertext = elgamal_encrypt(public_key, M, k)
    print(f"\nCiphertext: {ciphertext}")
    decrypted = elgamal_decrypt(private_key, public_key, ciphertext)
    print(f"\nDecrypted Message: {decrypted}")

if __name__ == "__main__":
    main()
```
